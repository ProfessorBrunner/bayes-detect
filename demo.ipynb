```python
prior_array = [[0.0,200.0],[0.0,200.0],[1.0,12.5],[2.0,9.0]]
#uniform prior for each parameter. Sets them up in order of [X,Y,A,R]
```

Parameters:
    -iterations
        -sets number of iterations, this doesn't actually matter because we run until convergence
    -samples
        -number of active samples for the nested sampling
        -increasing/decreasing affects speed
    -sample_method
        - "uniform"
            -uniform sampler
            -unusably slow
        - "metropolis"
            -metropolis hastings
        - "clustered_ellipsoidal"
            -clusters each iteration
            -selects a cluster and samples from within it
        - "new"
            -same as above except it clusters every 30 iterations

```python
from bayes_detect import sources
sources.run_source_detect(samples = 1200, iterations = 14000, sample_method = "new", prior = prior_array, noise_rms = noise, disp = 8.0, mode = "ipython")
```
